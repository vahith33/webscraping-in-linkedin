```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.edge.service import Service
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Set up the Edge driver
service = Service(EdgeChromiumDriverManager().install())
browser = webdriver.Edge(service=service)


    # Open LinkedIn
browser.get("https://www.linkedin.com")

# Wait for the page to load and the sign in options to appear
WebDriverWait(browser, 5).until(
    EC.presence_of_element_located((By.XPATH, "//a[contains(@href, 'login')]"))
)

# Click on the "Sign in with email" option
email_option = browser.find_element(By.XPATH, "//a[contains(@href, 'login')]")
email_option.click()

# Wait for the email and password fields to be present
WebDriverWait(browser, 10).until(
    EC.presence_of_element_located((By.ID, "username"))
)

# Find the username and password fields and enter your credentials
username = browser.find_element(By.ID, "username")
username.send_keys("ahamedvahith.k.aiml.2022@snsct.org")

password = browser.find_element(By.ID, "password")
password.send_keys("Pandya@33")

# Optionally, submit the form
password.send_keys(Keys.RETURN)

# Wait for a bit to see the result (you may want to check for successful login)
time.sleep(5)

# Use the search bar to search for jobs
search_bar = WebDriverWait(browser, 5).until(
    EC.presence_of_element_located((By.XPATH, "//input[contains(@placeholder, 'Search')]"))
)

search_bar.clear()  # Clear any existing text
search_bar.send_keys("Software Engineer")  # Change this to your desired job title
search_bar.send_keys(Keys.RETURN)  # Press Enter to submit the search

# Wait for the search results to load
time.sleep(5)

# Click on "See all job results in India"

see_all_jobs_link = WebDriverWait(browser, 5).until(
    EC.presence_of_element_located((By.XPATH, "//a[contains(text(), 'See all job results in India')]"))
)
see_all_jobs_link.click() 

time.sleep(5)

# Find job elements
job_titles=[]
company_names=[]
jobs = browser.find_elements(By.CLASS_NAME, "job-card-container")
for job in jobs:
    try:
        title = job.find_element(By.XPATH, "//span[strong and @aria-hidden='true']").text
        company = job.find_element(By.CLASS_NAME, "job-card-container__primary-description ").text
        job_titles.append(title)
        company_names.append(company)
    except Exception as e:
        print(f"Error: {e}")

# Save data to a DataFrame
df = pd.DataFrame({
    "Job Title": job_titles,
    "Company Name": company_names
})

print(df)

browser.quit()

# Save to CSV
df.to_csv("linkedin_jobs.csv", index=False)






```
